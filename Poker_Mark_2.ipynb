```python
full_deck = [{'value':i, 'suit':c}
for c in ['spades', 'clubs', 'hearts', 'diamonds']
for i in range(2,15)]
```

```python
import random as ran

#create function that draws first 5 cards from deck
def draw(deck):
    hand_drawn = []
    for i in range(0,5):
        hand_drawn.append(deck[i])
    return hand_drawn
```

```python
#function returns frequency dictionary
def card_freq_checker(hand,key):
    
        #extract card values from hand
        card_attributes = [card[key] for card in hand]
        
        #create output dictionary
        card_freq_dict = {}
        
        #for each new card value, add as a new key in dictionary with value 1
        #for each repeated card value, increment value for that key in the dictionary
        for value in card_attributes:
            if value in list(card_freq_dict.keys()):
                card_freq_dict[value] = card_freq_dict[value] + 1
            else:
                card_freq_dict[value] = 1
        
        #outputs a dictionary of each unique card value and its corresponding frequency
        return card_freq_dict
    
def flush_straight_checker(hand):

        suit_values_freq = list(card_freq_checker(hand,'suit').values())
        
        card_values = [card['value'] for card in hand]
        card_values.sort()
        
        if card_values==list(range(min(card_values),max(card_values)+1)) or card_values == [14,5,4,3,2]:
            if suit_values_freq==[5]:
                return 8
            else:
                return 4
            
        elif suit_values_freq==[5]:
            if card_values == [14,13,12,11,10]:
                return 9
            else:
                return 5
        else:
            return 0
        
        
        
#function returns the rank and value of a hand
def hand_rank(hand):
    
        rank_dict = {'rank':'','value':-1}
    
        flush_or_straight_value = flush_straight_checker(hand)
        
        if flush_or_straight_value == 9:
            rank_dict['rank'] = 'Royal Flush'
            rank_dict['value'] = 9
        elif flush_or_straight_value == 8:
            rank_dict['rank'] = 'Straight Flush'
            rank_dict['value'] = 8
        elif flush_or_straight_value == 5:
            rank_dict['rank'] = 'Flush'
            rank_dict['value'] = 5
        elif flush_or_straight_value == 4:
            rank_dict['rank'] = 'Straight'
            rank_dict['value'] = 4 
        
        else:
            
            #extract list of frequencies of unique cards of the hand using the frequency dictionary function
            card_val_freq_list = list(card_freq_checker(hand,'value').values())
            
            #sort the frequencies
            card_val_freq_list.sort(reverse=True)
            
            #each rank corresponds to a unique frequency list, check for each one
            #return that hand if they match
            if card_val_freq_list == [4,1]:
                rank_dict['rank']='Four-of-a-Kind'
                rank_dict['value'] = 7
            elif card_val_freq_list == [3,2]:
                rank_dict['rank']='Full House'
                rank_dict['value'] = 6
            elif card_val_freq_list == [3,1,1]:
                rank_dict['rank']='Three-of-a-Kind'
                rank_dict['value'] = 3
            elif card_val_freq_list == [2,2,1]:
                rank_dict['rank']='Two Pair'
                rank_dict['value'] = 2
            elif card_val_freq_list == [2,1,1,1]:
                rank_dict['rank']='One Pair'
                rank_dict['value'] = 1
            elif card_val_freq_list == [1,1,1,1,1]:
                rank_dict['rank']='High Card'
                rank_dict['value'] = 0
            
            #account for when there is no match
            else:
                rank_dict['rank']='Other Rank'
                
        return  rank_dict
```

```python
import itertools as iter
def best_hand(player_hand, river_hand):
    river_hand_combos = [list(cards) for cards in iter.combinations(river_hand, 3)]
    ranks = [hand_rank(player_hand + river_combo)['value'] for river_combo in river_hand_combos]
    return hand_rank(player_hand + river_hand_combos[ranks.index(max(ranks))])
```
